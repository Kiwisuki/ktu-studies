```python
elastingumai = [-2.1862, 0.0895, 0.35]
```

```python
def r(X, el=elastingumai):
    res = 1
    for i in range(len(X)):
        res *= X[i]**el[i]
        res *= (15**0.525 + 5)
    return res
```

```python
import numpy as np
```

```python
kainos = np.arange(900, 3000, 30)
```

```python
reklama = np.arange(10000, 600000, 1000)
```

```python
atstovai = np.arange(1, 30, 1)
```

```python
combos = np.array(np.meshgrid(kainos, reklama, atstovai)).T.reshape(-1,3)
```

```python
results = [r(X) for X in combos]
```

```python
import pandas as pd
```

```python
cols = ['Kainos', 'Reklama', 'Atstovai']
df = pd.DataFrame(combos, columns=cols)
df['Rezultatas'] = results
```

```python
df['Atstovai'] = df['Atstovai'].apply(lambda x: x*8000)
```

```python
df['Islaidos'] = df['Atstovai'] + df['Reklama']
```

```python
import seaborn as sns
```

```python
sns.scatterplot(df['Islaidos'], df['Rezultatas'])
```

```python
df['Rezultatas_scaled'] = df['Rezultatas'] * 125 * 2
df['Rezultatas_scaled'].mean()
```

```python
df['Pelnas'] = (df['Kainos'] - 1500) * 6000 * df['Rezultatas_scaled'] - df['Islaidos']
```

```python
sns.scatterplot(df['Rezultatas_scaled'], df['Pelnas'])
```

```python
sns.scatterplot(df['Islaidos'], df['Pelnas'])
```
