```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
train = pd.get_dummies(train.drop('Date', axis=1))
test = pd.get_dummies(test.drop('Date', axis=1))
```

```python
X = train.drop('Rented_Bike_Count', axis=1)
y = train['Rented_Bike_Count']

X_test = test.drop('Rented_Bike_Count', axis=1)
y_test = test['Rented_Bike_Count']

linear_model = LinearRegression(normalize=True).fit(X, y)

y_preds = linear_model.predict(X_test)
print(r2_score(y_test, y_preds))
print(mean_squared_error(y_preds, y_test, squared=False))
```

```python
ax = sns.scatterplot(x=y_test, y=y_preds)
ax.set(xlabel='Real', ylabel='Predictions')
plt.show()
```

```python
X = train.drop('Rented_Bike_Count', axis=1)
y = train['Rented_Bike_Count']
y = np.log1p(y)

X_test = test.drop('Rented_Bike_Count', axis=1)
y_test = test['Rented_Bike_Count']

linear_model = LinearRegression().fit(X, y)

y_preds = linear_model.predict(X_test)
y_preds = np.expm1(y_preds)

print(r2_score(y_test, y_preds))
print(mean_squared_error(y_test, y_preds, squared=False))
```

```python
ax = sns.scatterplot(x=y_test, y=y_preds)
ax.set(xlabel='Real', ylabel='Predictions')
plt.show()
```

```python
a = X['Wind_speed'] * X['Temperature']
b = X['Hour'] * X['Visibility']
c = X['Humidity'] * X['Temperature']
d = X['Hour'] * X['Temperature']

e = X_test['Wind_speed'] * X_test['Temperature']
f = X_test['Hour'] * X_test['Visibility']
g = X_test['Humidity'] * X_test['Temperature']
h = X_test['Hour'] * X_test['Temperature']

names = ['wind_temp', 'hour_vis', 'hum_temp', 'hour_temp']
trns = [a, b, c, d]
tst = [e, f, g, h]
```

```python
r2s =[]
mses = []

for n, i, j in zip(names, trns, tst):
    X = train.drop('Rented_Bike_Count', axis=1)
    X[n] = i
    y = train['Rented_Bike_Count']

    X_test = test.drop('Rented_Bike_Count', axis=1)
    X_test[n] = j

    y_test = test['Rented_Bike_Count']

    linear_model = LinearRegression().fit(X, y)

    y_preds = linear_model.predict(X_test)
    r2s.append(r2_score(y_test, y_preds))
    mses.append(mean_squared_error(y_test, y_preds, squared=False))
```
